```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("Laliga.csv")
df.head(10)
```

```python
df.info()
```

```python
 # 1. Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data
df = df.replace('-',0)
```

```python
# 2. Print all the teams which have started playing between 1930-1980.

demo = df[(df.Debut >= '1930') & (df.Debut <= '1980')]
demo = demo[['Team','Debut']]
print("Number of countries which started playing between 1930-1980",demo.shape)
demo.head(40)
```

```python
#3. Print the list of teams which came Top 5 in terms of points 

demo = df[['Team','Points']].copy()
demo['Points'] = demo['Points'].astype(float)
sorted_df = demo.sort_values(by='Points',ascending = False)
sorted_df.head(5)
```

```python
# 4 Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. 
# Using the same function, find the team which has maximum and minimum goal difference. (5 points)
# Goal_diff_count = GoalsFor - GoalsAgainst

def Goal_diff_count(goals_for,goals_Against):
    return (goals_for-goals_Against)

df['GoalsFor']= pd.to_numeric(df['GoalsFor'])
df['GoalsAgainst'] =  pd.to_numeric(df['GoalsAgainst'])


df['Goals_diff']=df.apply(lambda x: Goal_diff_count(x['GoalsFor'],x['GoalsAgainst']),axis=1)

print("team which has max goal difference \n ")
max = df['Goals_diff'].max()
print(df[df['Goals_diff']== max])
min = df['Goals_diff'].min()
print("team which has min goal difference")
print(df[df['Goals_diff']== min])

```

```python
# 5. Create a new column with name “Winning Percent” and append it to the data set (5 points)
    #Percentage of Winning = (GamesWon / GamesPlaye)*100

#If there are any numerical error, replace it with 0%

#Print the top 5 teams which has the highest Winning percentage

df['GamesWon'] = pd.to_numeric(df['GamesWon'])
df['GamesPlayed'] = pd.to_numeric(df['GamesPlayed'])
df['Winning_percentage']= (df['GamesWon']/df['GamesPlayed'])*100
df = df.fillna(0)
demo = df[['Team','Winning_percentage']].copy()

demo['Winning_percentage_float'] = demo['Winning_percentage'].astype(float)
demo1= demo.sort_values(by='Winning_percentage_float',ascending=False)
demo1.head(5)

```

```python
#6. Group teams based on their “Best position” and print the sum of their points for all positions (5 points)

df['Points']=pd.to_numeric(df['Points'])
demo1=df.groupby(by='BestPosition',axis=0)
demo1['Points'].sum()
```
